```python
import requests
import os
import time
```

```python
def download_image(url, file_name, path_to_save=None):
    with open(os.path.join(path_to_save,file_name), 'wb') as handle:
        response = requests.get(url, stream=True)

        if not response.ok:
            print(response)

        for block in response.iter_content(1024):
            if not block:
                break

            handle.write(block)
```

```python
path_to_save = "/shared-volume/UNATA STORAGE/docker_shared_files/data_root_own_laptop/data/image_text_datasets/google_conceptual_captions/images"
```

```python
download_track = []
for i in range(5000,7000):
    if not os.path.exists(os.path.join(path_to_save,f"{i}.jpg")):
        try:
            download_image(df[1].iloc[i], str(i)+".jpg",path_to_save)
        except Exception as e:
            print(e)
            pass
```

```python
def is_file_empty(file_path):
  """
  Checks if a file is empty (has 0 bytes).

  Args:
    file_path: The path to the file.

  Returns:
    True if the file is empty, False otherwise.
  """
  return os.path.getsize(file_path) == 0
```

```python
for i in range(5000):
    file_path = os.path.join(path_to_save,f"{i}.jpg")
    if os.path.exists(file_path) and is_file_empty(file_path):
        try:
            print(f">>> {file_path}\n")
            download_image(df[1].iloc[i], f"{i}.jpg",path_to_save)
            time.sleep(4)
        except:
            pass
```

```python
download_track = []
for i in range(7000):
    file_path = os.path.join(path_to_save,f"{i}.jpg")
    if os.path.exists(file_path) and is_file_empty(file_path):
        download_track.append([i,False])
    else:
        download_track.append([i,True])
```

```python
image_track = pd.DataFrame(download_track,columns=["index","is_image_available"])
```
