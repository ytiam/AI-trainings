### 1. For **discrete** random variables:

A **probability mass function (PMF)** is associated with the random variable ( X ), and it maps **real values** (that ( X ) can take) to **probabilities** in the range ([0, 1]), such that:

[
\sum_{x \in \text{Range}(X)} P(X = x) = 1
]

So:

* The function is ( P(X = x) )
* Domain: values that ( X ) can take (subset of ( \mathbb{R} ))
* Range: ([0, 1])

### 2. For **continuous** random variables:

A **probability density function (PDF)** is associated with ( X ), and it maps real numbers to **non-negative values**, but **not actual probabilities** directly.

[
f_X(x) \geq 0 \quad \text{and} \quad \int_{-\infty}^{\infty} f_X(x), dx = 1
]

So in this case:

* ( f_X(x) ) is the PDF
* Domain: ( \mathbb{R} )
* Range: ([0, \infty) ), as you said ✅

But to get actual probabilities, you **integrate** over an interval:

[
P(a \leq X \leq b) = \int_a^b f_X(x), dx
]

#### How and why the idea of likelihood came?

- Great question — and this shows you're thinking deeply about **why** core concepts like **likelihood** exist. Let's walk through the **intuition, motivation, and math** behind it step by step.

---

### 🧪 1. **Setting the Stage**: What’s the Problem?

Suppose you're a scientist, and you run an experiment — say, you flip a biased coin 10 times and get 7 heads and 3 tails.

Now you ask:

> “What value of the bias (probability of heads, say ( \theta )) best explains this observed data?”

This is **inverse thinking**:

* In probability, you usually ask:
  **Given** ( \theta ), what's the probability of the data?
* But here, you flip it:
  **Given** the data, what ( \theta ) is most plausible?

This leads to the concept of **likelihood**.

---

### 🎯 2. **What Is Likelihood?**

**Likelihood is a function of the parameter**, given observed data.

If:

* ( \theta ) is a parameter (e.g., bias of the coin),
* ( x ) is the observed data (e.g., 7 heads out of 10),

Then:
[
\mathcal{L}(\theta \mid x) = P(x \mid \theta)
]
The **same formula** as probability, but **different perspective**:

* In **probability**, ( x ) varies and ( \theta ) is fixed.
* In **likelihood**, ( x ) is fixed and ( \theta ) varies.

So the likelihood tells you:

> “How likely is it that this ( \theta ) could have generated the data I just saw?”

---

### 🧠 3. **Why Was It Introduced?**

It was formally introduced by **Ronald Fisher** in the 1920s, during the birth of **statistical inference**.

**Motivation**:

* He needed a way to estimate unknown parameters from real data.
* Probability couldn't help directly — it predicts data, not parameters.
* So he flipped the probability question around and created **likelihood** to “invert the problem.”

---

### 🧮 4. **Example: Binomial Likelihood**

Say you flipped a coin 10 times and saw 7 heads.
You model this as a **Binomial distribution**:
[
P(X = 7 \mid \theta) = \binom{10}{7} \theta^7 (1 - \theta)^3
]

Now define the **likelihood function**:
[
\mathcal{L}(\theta) = \binom{10}{7} \theta^7 (1 - \theta)^3
]

You can now:

* Plot it over different ( \theta ) values.
* Find the ( \theta ) that **maximizes** this — called the **Maximum Likelihood Estimate (MLE)**.

In this case, MLE = 0.7

---

| Concept     | Question                                                | Varies                 |
| ----------- | ------------------------------------------------------- | ---------------------- |
| Probability | “What is the probability of the data, given the model?” | Data varies            |
| Likelihood  | “How plausible is the model, given the data?”           | Model/parameter varies |
